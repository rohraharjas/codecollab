```python
import pandas as pd
import numpy as np
import sklearn as sk
```

```python
data = pd.read_csv('US_Accidents_March23.csv')
data.head()
```

```python
def percent_na(x):
    na_val = 0
    for val in x:
        if pd.isna(val):
            na_val += 1
    return na_val/len(x)

for feature in data.columns:
    null_per = percent_na(data[feature])
    print("%s: %5.4f"%(feature, null_per))
```
